```python
import csv
import matplotlib.pyplot as plt

def main():
    plt.rc('font',family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus']=False
    f = open('q3.csv','r',encoding='cp949')
    data = csv.reader(f,delimiter=',')
    header = next(data)
    row = next(data)
    male = []
    female = []
    year=[]
    for i in range(10):
        year.append(2013+i)
        male.append(int(row[i*2].strip().replace(',','')))
        female.append(int(row[i*2+1].strip().replace(',','')))
      
    plt.title("제주특별자치도 연도별 남,여 인구 수(2013~2022)")
    plt.xlabel('연도(년)')
    plt.ylabel('인구 수(명)')
    plt.bar(year,male,label='남자')
    plt.bar(year,female,label='여자')
    plt.xticks(year)
    plt.legend()
    plt.show()
    
    print(male)
    print(female)
    for i in range(len(male)):
        print(male[i]-female[i],end=" ")
    print()
    
    f.close()
    
if __name__=="__main__":
    main()
```
