```python
import csv
import random as r
import matplotlib.pyplot as plt
def makeCsv(size):
    filename = str(size)+'.csv'
    f=open(filename,'w',encoding="cp949")
    for i in range(size):
        if i == size-1:
            data = "%d"%(r.randrange(1,7))
        else:
            data = "%d\n"%(r.randrange(1,7))
        f.write(data)
    return f
def makeGraph(size,data,axe,c):
    count = []

    for row in data:
        count.append(int(row[0]))
    n, bins, patches = axe.hist(count,bins=6,color=c,alpha=0.7)
    axe.set_xlabel('눈')
    axe.set_ylabel('횟수')
    dice = []
    for i in range(1,7):
        dice.append(i)
    axe.set_xticks(dice)
    for k in range(len(patches)):
        axe.text(bins[k]+0.4, n[k], str(int(n[k])),fontsize=9, ha='center', va='top')  



def main():
    r.randrange(1,7)
    f=[]
    f.append(makeCsv(100))
    f.append(makeCsv(1000))
    f.append(makeCsv(10000))
    f.append(makeCsv(100000))
    for i in f:
        i.close()
    fread=[]
    fread.append(open('100.csv','r',encoding='cp949'))
    fread.append(open('1000.csv','r',encoding='cp949'))
    fread.append(open('10000.csv','r',encoding='cp949'))
    fread.append(open('100000.csv','r',encoding='cp949'))
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
    fig.suptitle('주사위 굴린 횟수 별 눈의 수')
    
    makeGraph(100,csv.reader(fread[0]),axes[0,0],'r')
    axes[0,0].set_title("100")
    makeGraph(1000,csv.reader(fread[1]),axes[0,1],'b')
    axes[0,1].set_title("1000")
    makeGraph(10000,csv.reader(fread[2]),axes[1,0],'g')
    axes[1,0].set_title("10000")
    makeGraph(100000,csv.reader(fread[3]),axes[1,1],'y')
    axes[1,1].set_title("100000")
#     plt.subplots(constrained_layout=True)
    plt.rc('font',family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus']=False
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    fig.tight_layout()
    plt.show()
    
    for i in fread:
        i.close()
if __name__=="__main__":
    main()
```
