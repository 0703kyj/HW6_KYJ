```python
import csv
import matplotlib.pyplot as plt

def main():
    plt.rc('font',family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus']=False
    f = open('q4.csv','r',encoding='cp949')
    data = csv.reader(f,delimiter=',')
    
#     trans = []
    enter = {}
    out = {}
    total = {}
    
    for row in data:
        trans = row[0].strip()
        enter[trans] = int(row[1].strip())
        out[trans] = int(row[2].strip())
        enter[trans]+=int(row[3].strip())
        out[trans]+=int(row[4].strip())
        total[trans] = enter[trans]+out[trans]
        if trans == '역삼':
            print("역삼 승차 수 :",enter[trans])
        elif trans == "구로디지털단지":
            print("구로 하차 수 :",out[trans])
        
    sort_enter = sorted(enter.items(), key = lambda item: item[1],reverse=True)
    sort_out = sorted(out.items(), key = lambda item: item[1],reverse=True)
    sort_total = sorted(total.items(), key = lambda item: item[1],reverse=True)
    graph_enter = []
    graph_out = []
    graph_total = []
    for i in range(30):
        graph_enter.append(sort_enter[i])
        graph_out.append(sort_out[i])
        graph_total.append(sort_total[i])
#     for i in graph_total:
#         print(i)    
    fig, axes = plt.subplots(nrows=3, ncols=1)
    fig.suptitle('출근 시간대(07시~09시) 지하철 역 별 승하차 승객 수',y=6.1)
    fig.tight_layout()
    
    plt.title("2022 temp")
    for i in range(3):
        axes[i].set_xlabel('승객수(명)')
        axes[i].set_ylabel('역')
#     plt.tight_layout()
    plt.subplots_adjust(top=6,hspace=0.1)
    for i in graph_enter:
        axes[0].set_title("승차")
        axes[0].barh(i[0],i[1], height=0.5)
    for i in graph_out:
        axes[1].set_title("하차")
        axes[1].barh(i[0],i[1], height=0.5)
    for i in graph_total:
        axes[2].set_title("승하차")
        axes[2].barh(i[0],i[1], height=0.5)    
    plt.show()
    
    print("승차 :",graph_enter[0])
    print("하차 :",graph_out[0])
    print("승하차 :",graph_total[0])
    
    f.close()
if __name__=="__main__":
    main()
```
