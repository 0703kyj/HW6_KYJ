```python
import csv
import matplotlib.pyplot as plt
def main():
    plt.rc('font',family = 'Malgun Gothic')
    plt.rcParams['axes.unicode_minus']=False
    f=[]
    f.append(open('korea.csv','r',encoding='cp949'))
    f.append(open('seoul.csv','r',encoding='cp949'))
    f.append(open('busan.csv','r',encoding='cp949'))
    f.append(open('daejon.csv','r',encoding='cp949'))
    f.append(open('jeju.csv','r',encoding='cp949'))
    data = []
    loc = ['전국','서울','부산','대전','제주']
#     loc = ['korea','seoul','busan','daejeon','jeju']
    temp = []
    month = []
    for i in range(1,13):
        month.append(i)
#     print(csv.reader(f[0],delimiter=','))
    for i in range(len(f)):
        data.append(csv.reader(f[i],delimiter=','))
#         tmp_loc = str(next(data[i]))
#         loc.append(tmp_loc[-4:-2])
        header = next(data[i])
        tmp_list = []
        for row in data[i]:
            tmp_list.append(float(row[2]))
        temp.append(tmp_list)
    plt.title("2022 지역별 월평균 기온")
    plt.xlabel('month(월)')
    plt.ylabel('temperature(°C)')
    for i in range(len(loc)):
        if loc[i] == '전국':
            plt.plot(month,temp[i],'-',label=loc[i])
        else:    
            plt.plot(month,temp[i],'x',label=loc[i])
    plt.xticks(month)
    plt.legend()
    plt.show()
    
    for i in range(len(temp)):
        print(loc[i],end=" ")
        print(temp[i])
    
    for i in range(1,len(temp)):
        print(loc[i],end=" ")
        count=0
        for j in range(len(temp[0])):
            diff=round(temp[i][j]-temp[0][j],1)
            print(diff,end=" ")
            if diff<0:
                count+=1
        print("/ 개수 : ",count)
    
    for i in f:
        i.close()
if __name__=="__main__":
    main()
```
